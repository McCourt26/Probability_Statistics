```python
import numpy as np 
import matplotlib.pyplot as plt
t = np.arange(0.0, 2.0, 0.01) #This reprensnets range 
s = np.sin(2*np.pi*t)

plt.plot(t, s)
plt.title(r'$\alpha_i > \beta_i$', fontsize=20)
plt.text(1, -0.6, r'$\sum_{i=0}^\infty x_i$', fontsize=20) #In order to change font size and font 
plt.text(0.5, 0.6, r'$\mathcal{A}\mathrm{sin}(2 \omega t)$',fontsize=15)
plt.xlabel('time (s)')
plt.ylabel('volts (mV)')
plt.show()
```

```python
import matplotlib.pyplot as corndog #You can change the name as I did here with "corndog"
import numpy as np #This is a library of math fuctions 

# Data for plotting
x = np.arange(-10,10,0.001)

y = (10 * x) + 7 #This is telling computer what function is 
_, ax = corndog.subplots()
ax.plot(x, y) #This is telling computer there will be a line 



ax.set(xlabel='time (s)', ylabel='voltage (mV)', #X and Y label
       title='About as simple as it gets, folks')
corndog.show()
```
